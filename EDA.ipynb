# Car data analysis

the aim of this notebook is to analyze the data coming from the 3 cameras mounted on the front of the car

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

# 1st scenario backward cases

```python
# reading data
root_folter = './data/'
data_1st_scenario_fwd = root_folter+'data_1st_scenario_backward/driving_log.csv'
df_1st_bwd = pd.read_csv(data_1st_scenario_fwd, names=['center image path', 'left image path', 'right image path', 
                                                'steering angle', 'Throttle value','Break value',
                                               'speed'])
df_1st_bwd.head()
```

```python
plt.figure(figsize=(15,4))
plt.subplot(121)
df_1st_bwd['steering angle'].hist(bins=20);
plt.title('complete steering data histogram', fontsize=20)
plt.xlabel('steering angle values')
plt.subplot(122)
df_1st_bwd['steering angle'].hist(bins=20);
plt.ylim([0,500])
plt.title('steering data histogram zoom', fontsize=20)
plt.xlabel('steering angle values')
plt.xticks(np.linspace(-1,1,21), rotation=70)
plt.show()
```

```python
df_1st_bwd['steering angle'].describe()
```

### visualizing one image

```python
img_no = 0
fig=plt.figure(figsize=(15,5))
img_path = df_1st_bwd['center image path'][img_no]
img = cv2.imread(img_path)
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.subplot(131)
plt.imshow(rgb_img);
plt.xlabel(img_path.split('/')[-1])
plt.title(f'image shape: {rgb_img.shape}')

img_path = df_1st_bwd['left image path'][img_no]
img = cv2.imread(img_path)
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.subplot(132)
plt.imshow(rgb_img);
plt.xlabel(img_path.split('/')[-1])
plt.title(f'image shape: {rgb_img.shape}')

img_path = df_1st_bwd['right image path'][img_no]
img = cv2.imread(img_path)
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.subplot(133)
plt.imshow(rgb_img);
plt.xlabel(img_path.split('/')[-1])
plt.title(f'image shape: {rgb_img.shape}')
fig.tight_layout()
```

# Filter out some data
in the three scenarios we have data that skews to ranges around steering angles of zero, which it is expected since the driver has driven around the track trying his best to be in the central line. To avoid this we will cut all data around zero to a number of 500 at most.

```python
high_limit = 500
# 1st scenario backward
number_to_drop = len(df_1st_bwd[df_1st_bwd['steering angle']==0.0])-high_limit
df_1st_bwd.drop(df_1st_bwd[df_1st_bwd['steering angle']==0.0].sample(number_to_drop).index, inplace=True)
# 1st scenario forward
number_to_drop = len(df_1st_fwd[df_1st_fwd['steering angle']==0.0])-high_limit
df_1st_fwd.drop(df_1st_fwd[df_1st_fwd['steering angle']==0.0].sample(number_to_drop).index, inplace=True)
# 2nd scenario forward
number_to_drop = len(df_2nd_fwd[df_2nd_fwd['steering angle']==0.0])-high_limit
df_2nd_fwd.drop(df_2nd_fwd[df_2nd_fwd['steering angle']==0.0].sample(number_to_drop).index, inplace=True)
```

```python
# 1st scenario backward
plt.figure(figsize=(15,4))
plt.subplot(121)
df_1st_bwd['steering angle'].hist(bins=20);
plt.title('complete steering data histogram', fontsize=20)
plt.xlabel('steering angle values')
plt.subplot(122)
df_1st_bwd['steering angle'].hist(bins=20);
plt.ylim([0,500])
plt.title('steering data histogram zoom', fontsize=20)
plt.xlabel('steering angle values')
plt.xticks(np.linspace(-1,1,21), rotation=70)
plt.show()
```

```python
# 1st scenario forward
plt.figure(figsize=(15,4))
plt.subplot(121)
df_1st_fwd['steering angle'].hist(bins=20);
plt.title('complete steering data histogram', fontsize=20)
plt.xlabel('steering angle values')
plt.subplot(122)
df_1st_fwd['steering angle'].hist(bins=20);
plt.ylim([0,500])
plt.title('steering data histogram zoom', fontsize=20)
plt.xlabel('steering angle values')
plt.xticks(np.linspace(-1,1,21), rotation=70)
plt.show()
```

```python
# 2nd scenario forward
plt.figure(figsize=(15,4))
plt.subplot(121)
df_2nd_fwd['steering angle'].hist(bins=20);
plt.title('complete steering data histogram', fontsize=20)
plt.xlabel('steering angle values')
plt.subplot(122)
df_2nd_fwd['steering angle'].hist(bins=20);
plt.ylim([0,500])
plt.title('steering data histogram zoom', fontsize=20)
plt.xlabel('steering angle values')
plt.xticks(np.linspace(-1,1,21), rotation=70)
plt.show()
```

# Joining the three dataframes in one
